This notebook can run on [TACC Visualization Portal](https://vis.tacc.utexas.edu/) to install [OpenBLAS](http://www.openblas.net/), [Theano](http://deeplearning.net/software/theano/), and [Lasagne](http://lasagne.readthedocs.org/en/latest/).

### Install OpenBLAS

```python
%%bash
cd $HOME
git clone git://github.com/xianyi/OpenBLAS
cd OpenBLAS
make clean
make FC=gfortran
make PREFIX=$HOME/.local install
```

### Install Theano

```python
%%bash
cd $HOME
git clone git://github.com/Theano/Theano.git
cd Theano
export BLAS=$HOME/.local/lib/libopenblas.a
export LAPACK=$HOME/.local/lib/libopenblas.a
python setup.py develop --user
```

### Install Lasagne

```python
%%bash
cd $HOME
git clone https://github.com/Lasagne/Lasagne
cd Lasagne
python setup.py develop --user
```

### Theano Environment Variables

The following are environment variables necessary to use the GPU in Theano. See [Theano configuration](http://deeplearning.net/software/theano/library/config.html) and [Using the GPU](http://deeplearning.net/software/theano/library/config.html).

```python
%%writefile .theanorc
[global]
device=gpu
floatX=float32

[blas]
ldflags=-lopenblas

[cuda]
root=/opt/apps/cuda/7.0

[nvcc]
fastmath=True
```
